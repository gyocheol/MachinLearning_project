```python
import pandas as pd
import numpy as np

# ...없애기
# pd.options.display.max_columns = None
# pd.options.display.max_rows = None
```

```python
df_a = pd.read_csv('./datasets/2015.csv')
for i in range(2016,2023):
    year = i
    df_b = pd.read_csv(f'./datasets/{year}.csv')
    df_a = pd.concat([df_a, df_b])
df_a
```

```python
df_a['월'] = df_a['월'].replace({'1월':'1', '2월':'2', '3월':'3', '4월':'4', '5월':'5', '6월':'6', 
                               '7월':'7', '8월':'8', '9월':'9', '10월':'10', '11월':'11', '12월':'12' })
df_a
```

```python
df = df_a.iloc[:, :-1]
df
```

```python
df.to_csv("15~22년월별전력량.csv", index=False)
```
