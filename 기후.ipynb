```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./datasets/기상.csv', encoding='cp949')
df
```

```python
df.drop('지점', axis=1, inplace=True)
```

```python
df = df.rename(columns={'지점명':'시도'})
df
```

```python
df.loc[df['시도']=='서울','시도']='서울특별시'
df.loc[df['시도']=='부산','시도']='부산광역시'
df.loc[df['시도']=='대구','시도']='대구광역시'
df.loc[df['시도']=='인천','시도']='인천광역시'
df.loc[df['시도']=='광주','시도']='광주광역시'
df.loc[df['시도']=='대전','시도']='대전광역시'
df.loc[df['시도']=='울산','시도']='울산광역시'
```

```python
df
```

```python
import datetime
df['일시2'] = np.zeros(595)
for i in range(len(df['일시'])):
    format = '%b-%y'
    df['일시2'][i] = datetime.datetime.strptime(df['일시'][i],format)
    
df['년도'] = pd.DatetimeIndex(df['일시2']).year
df['월'] = pd.DatetimeIndex(df['일시2']).month
df
```

```python
df = df.drop('일시', axis=1)
df
```
